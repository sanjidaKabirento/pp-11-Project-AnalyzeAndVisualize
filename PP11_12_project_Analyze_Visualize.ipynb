```python
df = pd.read_csv("student_data.csv")
df.head()
df.sample(5)
```

```python
##Create a pie chart showing the distribution of students by department.
district_counts = df['home_district'].value_counts()

print(district_counts)
```

```python
## Display students who have both SSC and HSC GPAs above 4.
filtered_students = df[(df['ssc_gpa'] > 4) & (df['hsc_gpa'] > 4)]
print(filtered_students[["first_name",	"last_name"	]])
```

```python
## Calculate the total number of students with an overseas study experience.
study_abroad_country_count= df["study_abroad_country"].value_counts()

print(study_abroad_country_count)
```

```python
##Replace missing values in the "cgpa" column with the average CGPA.
df.isnull().sum()
df_fillna=df.fillna({"cgpa":df["cgpa"].mean()})
df_fillna.isnull().sum()
```

```python
##Create a pie chart showing the distribution of students by department.
import matplotlib.pyplot as plt
dept=df["department"].value_counts()
print(dept)
dept1=["Medicine","Computer Science","Business","Arts","Engineering"]
dept_array = dept.values
dept_list = dept.values.tolist()
color=['blue',"Green","skyblue","red","pink"]
colors = plt.cm.get_cmap('tab10', len(dept))
plt.pie(dept_list,colors= colors.colors,autopct='%1.1f%%',startangle=140)
plt.axis('equal')
plt.title('Department')
plt.show()
```
